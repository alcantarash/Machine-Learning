```python
resultRandom = []
```

```python
def checkifRandom(vetor, index, y): #função para checar se já existe no vetor
    x = 0
    for i in vetor[index]:
        if int(i) == y:
            x = 1
            break
    if x == 1:
        checkifRandom(users[0], index, randint(0, len(vetor)))
    else:     
        print('The recommendation for the user', index, 'is the article' , y)
        resultRandom.append(y) #coloca no result o resultado
    return
```

```python
def suggest_randomly(users): #função de sugestão aleatória
    len_user = len(users[0])
    for index, x in enumerate(users[0]):
        y = randint(0, len_user)
        checkifRandom(users[0], index, y)
    return
```

```python
suggest_randomly(users)
```

# Most Popular

```python
articles = {}
for k in range(len(raw_data)):
    articles[k+1] = 0

#Vetor para armazenar quantas vezes um artigo foi lido    
for i in range(len(users)):
    for j in users[0][i]:
        if int(j) != 0:
            articles[int(j)] = articles[int(j)] + 1

#Ordenando o vetor de artigos lidos            
aux = []
articles_sorted = sorted(articles, key=articles.get, reverse=True)# "ranking" dos artigos lidos
for r in articles_sorted:
    print r, articles[r]
    aux.append(r)
```

```python
#Verifica se o usuário já leu o artigo e recomenda o próximo mais popular
def checkReadMostPopular(vetor, index, position):
    read = 0
    for i in vetor[index]:
        if int(i) == aux[position]:
            read = 1
            break
    if read == 1:
        position = position + 1
        checkReadMostPopular(users[0], index, position)
    else:
        vetor[index].append(aux[position])
        print('The recommendation for the user', index, 'is the article' , aux[position])

#Função para recomendar o artigo mais popular        
def suggest_mostPopular(users):
    for index, x in enumerate(users[0]):
        checkReadMostPopular(users[0], index, 0)
    return    
```

```python
suggest_mostPopular(users)
```

# Content-based

```python
citations = pd.read_csv('citations.dat',  header=None)
```

```python
citations[0] = citations[0].str.split(' ')
```

```python
#Função para pegar os artigos citados nos artigos que o usuário leu
def otherArticles(vetArtLidos):
    listarticles = []
    for i in vetArtLidos:
            for j in citations[0][int(i)]:
                if int(j) != 0:
                    listarticles.append(j)
    return list(set(listarticles))
        
#Função para checar se o artigo recomendado já foi lido pelo usuário
def checkReadBasedContent(vetor, index, articles, position):
    read = 0
    for i in vetor[index]:
        if int(i) == articles[position]:
            read = 1
            break
    if read == 1:
        position = position + 1
        checkReadBasedContent(users[0], index, articles, position)
    else:
        vetor[index].append(articles[position])
        print('The recommendation for the user', index, 'is the article' , articles[position])

    return

#Função para recomendar o artigo baseado em conteúdo
def suggest_basedContent(users):
        for index, x in enumerate(users[0]):
            userArticles = otherArticles(x)
            checkReadBasedContent(users[0], index, userArticles, 0)
            
        return
```
